```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(101)
```

```python
array=np.random.randint(10,50,(10,5))
df=pd.DataFrame(data=array,index="A B C D E F G H I J".split(),columns="val1 val2 val3 val4 val5".split())
df
```

```python
#to return values greater than 20
df>20
```

```python
df[df>20]
```

```python
#to do columnwise 
#to get elements less than 20 from val1
df["val1"]<20
```

```python
df[["val1"]]<20
```

```python
df[df["val1"]<20]
```

```python
#to get val2 column
df[df["val1"]<20]["val2"]
```

```python
df[df["val1"]<20][["val2"]]
```

```python
#to return val5 next to val2
df[df["val1"]<20][["val2","val5"]]
```

```python
df
```

```python
#to use AND (&) operator
#To retrive numbers in val1 column greater than 20 and val4 column less than 18 in the dataframe
df[(df["val1"]>20)&(df["val4"]<18)]
```

```python
#To use OR (|) operator
#to return values less than 35 in the whole dataframe and those greater than 20 in column 5
df[(df<35)|(df["val5"]>20)]
```

```python
df
```

```python
#to return values greater than 25 from val2 and return corresponding columns val5 and val3
df.loc[df["val2"]>25,["val2","val3","val5"]]
```

```python
df.loc[df.val2>25,["val2","val3","val5"]]
```

```python
df1 = pd.DataFrame({'employee': ['Julia', 'Marie', 'Adam', 'Nicole'],
                     'department': ['Data Science', 'Web Development', 'Data Science', 'Cyber Security'],
                     'Year': ['2005', '2008', '2011', '2002']})
   
df2 = pd.DataFrame({'employee': ['Nicole', 'Adam', 'Julia', 'Marie'],
                          'country': ['Canada', 'England', 'USA', 'Germany'],
                          'salary': ['22000', '16000', '20000', '17500']})
```

```python
df3=pd.merge(df1,df2)
df3
```

```python
#to perform merge() if we dont have 2 data frames with common variable names
```

```python
df5
```

```python
pd.merge(df3,df5,left_on="employee",right_on="member")
```

```python
pd.merge(df3,df5,left_on="employee",right_on="member",how="outer")
```

```python
pd.merge(df3,df5,left_on="employee",right_on="member",how="left")
```

```python
pd.merge(df3,df5,left_on="employee",right_on="member",how="right")
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2,how="inner")
```

```python
df1.join(df2,how="outer")
```

```python
df1.join(df2,how="left")
```

```python
df1.join(df2,how="right")
```

```python
#107 Pivot Tables in Pandas Library
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df=sns.load_dataset("titanic")
df.head()
```

```python
df.groupby("sex")[["survived"]].mean()
```

```python
df.groupby(["sex","embark_town"])[["survived"]].mean()
```

```python
df.groupby(["sex","embark_town"])[["survived"]].mean().unstack()
```

```python
df.pivot_table("survived",index="sex",columns="embark_town")
```

```python
df.pivot_table("survived",index="sex",columns="embark_town",aggfunc=["sum","mean","std"])
```

```python
df.pivot_table("survived",index=["sex","class"],columns="embark_town",aggfunc=["sum","mean","std"])
```
